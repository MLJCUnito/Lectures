```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=40) 
plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
#histtype : {'bar', 'barstacked', 'step', 'stepfilled'}
#if density = True the sum of the histograms is normalized to 1.
```

```python
#If you would like to simply compute the histogram
#(that is, count the number of points in a given bin) and not display it,
#the np.histogram() function is available
data = np.random.randn(1000)
counts, bin_edges = np.histogram(data, bins=5)
print(counts)
```

### basic barplot

```python
import numpy as np
import matplotlib.pyplot as plt
 
# Make a fake dataset:
height = (33, 12, 49, 18, 45)
bars = ['A', 'B', 'C', 'D', 'E']
bars_pos = np.arange(len(bars))       
 
# Create bars
plt.bar(bars_pos, height, color = 'r')
 

plt.xticks(bars_pos, bars)

plt.ylabel('height');
```

### choose color of barplots and control its borders

```python
import numpy as np
import matplotlib.pyplot as plt 
height = [33, 12, 49, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')
bars_pos = np.arange(len(bars))
plt.bar(bars_pos, height, color=['black', 'red', 'green', 'blue', 'cyan']);
plt.xticks(bars_pos, bars);
```

### example

```python
import numpy as np
import matplotlib.pyplot as plt


N = 5
menMeans = (50, 69, 64, 55, 58)      #set
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups     arange return evenly spaced values within a given interval.
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, menMeans, width, yerr=menStd)
p2 = plt.bar(ind, womenMeans, width,
             bottom=menMeans, yerr=womenStd)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 120, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'));

```

### Annotations

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c, s = np.cos(x), np.sin(x)
plt.figure(figsize=(16, 6), dpi=80)
plt.plot(x, c, color="blue", linewidth=2.5, linestyle="-")
plt.plot(x, s, color="red",  linewidth=2.5, linestyle="-")
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])            #latex
plt.yticks([-1, 0, +1],
           [r'$-1$', r'$0$', r'$+1$'])


# Legend:
plt.plot(x, c, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(x, c, color="red",  linewidth=2.5, linestyle="-", label="sine")
plt.legend(loc='upper left')

# Annotations:
t = 2*np.pi/3
plt.plot([t, t], [0, np.cos(t)], color='blue', linewidth=2.5, linestyle="--")
plt.scatter([t, ], [np.cos(t), ], 50, color='blue')
plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$',
             xy=(t, np.cos(t)), xycoords='data',
             xytext=(-90, -50), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->"));

plt.plot([t, t],[0, np.sin(t)], color='red', linewidth=2.5, linestyle="--")
plt.scatter([t, ],[np.sin(t), ], 50, color='red')
plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
             xy=(t, np.sin(t)), xycoords='data',
             xytext=(+10, +30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->"));

```

### Example: 3D plot

```python
#thanks to add-for!!!
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X ** 2 + Y ** 2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.hot)
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.cm.hot)
ax.set_zlim(-2, 2)

plt.show()
```
